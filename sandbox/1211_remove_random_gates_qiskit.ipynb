Goal: Figure out how to remove random gates from a qiskit quantum circuit.

Let's consider `qft_4` as an example circuit:

```python
qft_4_path = "../feynman/benchmarks/qasm/qft_4.qasm"

from qiskit import QuantumCircuit
qft_4_circ = QuantumCircuit.from_qasm_file(qft_4_path)
```

```python
qft_4_circ.draw(output="mpl")
```

```python
del qft_4_circ.data[1]
```

```python
qft_4_circ.draw(output="mpl")
```
