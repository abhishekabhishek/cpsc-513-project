```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
qaoa_eq_df = pd.read_csv('results/mqtbench_qaoa_min_2_max_20_step_1_eq_checking_results.csv')
qaoa_non_eq_df = pd.read_csv('results/mqtbench_qaoa_min_2_max_20_step_1_non_eq_checking_results.csv')

qft_eq_df = pd.read_csv('results/mqtbench_qft_min_2_max_16_step_1_eq_checking_results.csv')
qft_non_eq_df = pd.read_csv('results/mqtbench_qft_min_2_max_12_step_1_non_eq_checking_results.csv')

qfte_eq_df = pd.read_csv('results/mqtbench_qftentangled_min_2_max_20_step_1_eq_checking_results.csv')
qfte_non_eq_df = pd.read_csv('results/mqtbench_qftentangled_min_2_max_20_step_1_non_eq_checking_results.csv')

vqe_eq_df = pd.read_csv('results/mqtbench_vqe_min_2_max_21_step_1_eq_checking_results.csv')
vqe_non_eq_df = pd.read_csv('results/mqtbench_vqe_min_2_max_20_step_1_non_eq_checking_results.csv')
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(32, 12))
for label, df in zip(
    ['qaoa', 'qft', 'qftentangled', 'vqe'],
    [qaoa_eq_df, qft_eq_df, qfte_eq_df, vqe_eq_df]):
    ax1.plot(df['n_qubits'].values, df['check_time'].values,
            marker='o',
            linestyle='--',
            label=label)
    
for label, df in zip(
    ['qaoa', 'qft', 'qftentangled', 'vqe'],
    [qaoa_non_eq_df, qft_non_eq_df, qfte_non_eq_df, vqe_non_eq_df]):
    ax2.plot(df['n_qubits'].values, df['check_time'].values,
            marker='o',
            linestyle='--',
            label=label)
    
ax1.set_xlabel('No. of qubits', fontsize=30)
ax1.set_ylabel('Runtime (seconds)', fontsize=30)
ax1.set_yscale('log')
ax1.tick_params(axis='both', labelsize=30)
ax1.legend(fontsize=30)
ax1.grid(True)
ax1.set_title('Equivalence checking', fontsize=30)

ax2.set_xlabel('No. of qubits', fontsize=30)
ax2.set_yscale('log')
ax2.tick_params(axis='both', labelsize=30)
ax2.grid(True)
ax2.set_title('Non-equivalence checking', fontsize=30)

fig.suptitle('Runtimes to perform equivalence and non-equivalence checking for \ndifferent quantum circuits with increasing no. of qubits\n', fontsize=30)
fig.tight_layout()
```
